```python
#self work
data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Finance', 'Finance', 'IT'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Salary': [50000, 52000, 60000, 58000, 65000, 62000, 61000],
    'Experience': [2, 3, 5, 4, 7, 6, 5]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.groupby(['Department', 'Experience'])['Salary'].mean()
```

```python
df.groupby('Department').agg({
    'Salary': ['mean', 'max'],
    'Experience': ['mean', 'min']
})
```

```python
df.groupby('Department').count()
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
df.groupby('Department')['Salary'].max()
```

```python
df.groupby('Department')['Salary'].mean(numeric_only=True)
```
