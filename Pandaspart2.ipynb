```python
# Create a function called quality which will check quality of the tip if tip is greater than 25% of bill quality is good otherwise bad.
```

```python
def quality(tip,total_bill):
    if tip/total_bill>0.25:
        return "Generous"
    else:
        return "Other"
    
    
```

```python
#apply :pandas
#Numpy: vectorise
# df['tip quality']=df[['total_bill','tip']].apply(Function name)# this is Pandas function.
df['tip_quality']=np.vectorize(quality)(df['total_bill'],df['tip'])# faster than 'apply' ,time complecity is less and this is Numpy function .
```

```python
df['tip_quality2']=df[['tip','total_bill']].apply(lambda df :quality(df['total_bill'],df['tip']),axis=1)
```

# Alternative with custom apply() call

```python
def cleanup(name):
    name=name.replace(";","")
    name=name.strip()
    name=name.capitalize()
    return name
messy_names.apply(cleanup)
```
